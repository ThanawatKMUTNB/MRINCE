```python
import pandas as pd
import numpy as np
```

```python
csv = pd.read_csv('../orders-2022-06-20-00-16-07.csv')
```

```python
Number = 8421

customer_info = csv.loc[csv['No.'] == Number].iloc[0]
info = f'''
ชื่อลูกค้า : {customer_info['Customer Name']}
สถานะ : {customer_info['Status']}
โทรศัพท์ : {customer_info['Customer Phone']}
อีเมล : {customer_info['Customer Email']}
ที่อยู่ : {customer_info['Customer Address']}
'''
print(info)
```

```python
order_list = csv.loc[csv['No.'] == Number][['Item Name', 'Item Qty']]
order_list = order_list.assign(GetItem=4)
order_list
```

```python
def get_report(DF : pd.DataFrame) -> pd.DataFrame:
    data = pd.DataFrame(columns=['Item Name', 'Status', 'Item Qty'])
    for index, row in DF.iterrows():
        n = row['Item Qty'] - row['GetItem']
        if n > 0: # ขาด
            df = pd.DataFrame(
                {
                    'Item Name' : [row['Item Name']],
                    'Status' : ['ขาด'],
                    'Item Qty' : [n]
                }
            )
            # print(f"{row['Item Name']} ขาด")
        elif n < 0: # เกิน
            df = pd.DataFrame(
                {
                    'Item Name' : [row['Item Name']],
                    'Status' : ['เกิน'],
                    'Item Qty' :[-n]
                }
            )
        else:
            continue
        data = pd.concat([data,df], ignore_index=True)
            # print(f"{row['Item Name']} เกิน")
    return data

get_report(order_list)
```
